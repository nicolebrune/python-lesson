```python
import urllib.request
import pandas as pd
#urllib liegt bereits auf dem PC zusammen mit vielen weiteren Bibliotheken (über Anaconda heruntergeladen).
#urllib ist das Package, womit wir die Daten herunterladen. Oft sind Packages in Module aufgeteilt.
#pandas erhält ein Alias.
```

```python
open_apc_url = "https://raw.githubusercontent.com/OpenAPC/openapc-de/master/data/fuberlin/APC_FU_Berlin_2015.csv"
```

```python
apc_file = "openapc.csv"
#Wird grundsätzlich in dem Ordner abgelegt, wo sich auch das Skript befindet.
#Hier lassen sich auch Pfadangaben inkludieren, z.B. C/users/xxx/Downloads/openapc.csv
```

```python
urllib.request.urlretrieve(open_apc_url, apc_file)
#Nach dem Modul request folgt urlretrieve als Funktion mit 2 Argumenten.
```

```python
apcs = pd.read_csv(apc_file)
#apcs = article processing charges
#read_csv = Funktion von pandas
#Hier wird das CSV eingelesen und ein pandas-DataFrame kreiert.
```
